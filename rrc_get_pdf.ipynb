```python
import pickle
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import contextlib
from selenium.webdriver.support.ui import Select
import urllib
import glob
import os
from random import randint
from time import sleep

import chromedriver_binary
```

```python
with open('pickle/midland_apis.pickle', 'rb') as to_read: #run scrape_data to generate
    midland_apis = pickle.load(to_read)
```

```python
def initialize_driver(dir_path):
    options = webdriver.ChromeOptions()
    profile = {"plugins.plugins_list": [{"enabled": False, "name": "Chrome PDF Viewer"}], # Disable Chrome's PDF Viewer
               "download.default_directory": dir_path , "download.extensions_to_open": "applications/pdf"}

    options.add_experimental_option("prefs", profile)
    driver = webdriver.Chrome(chrome_options=options)  # Optional argument, if not specified will search path.
    return driver
```

```python
def completion_query(api,driver,dir_path):
    search_url = 'http://webapps.rrc.texas.gov/CMPL/publicSearchAction.do'
    api = str(api)
    print(api)
    api_nostate = str(api)[2:]
    
    #search for comp reports with given API number
    driver.get(search_url)
    api_box = driver.find_element_by_name("searchArgs.apiNoHndlr.inputValue")
    api_box.send_keys(api_nostate)
    search = driver.find_element_by_xpath("/html/body/table[6]/tbody/tr/td/form/table/tbody/tr[4]/td/table/tbody/tr[5]/td/table/tbody/tr/td/input[1]")
    search.click()

    
    #get list of completion report rows
    comp_table = driver.find_element_by_class_name("DataGrid")
    completion_reports = comp_table.find_elements_by_tag_name("tr")
    
    #select each report and grab pdf
    report_link_list = []
    for report in completion_reports[3:]:
        completion = report.find_elements_by_tag_name('td')[0]
        completion_link = completion.find_element_by_tag_name('a')
        report_link = str(completion_link.get_attribute('href'))
        report_link_list.append(report_link)
        
    report_count = 0
    for report in report_link_list:
        driver.get(report)
        w2_link = driver.find_element_by_xpath("/html/body/table[6]/tbody/tr/td/form/table/tbody/tr/td/table[2]/tbody/tr[2]/td[1]/table/tbody/tr[2]/td/table/tbody/tr[2]/td[3]/a")
        w2_link = str(w2_link.get_attribute('href'))
        driver.get(w2_link)
        sleep(randint(1,5))
        report_count += 1
        rename_file(get_last(dir_path),api+'-'+str(report_count))
        return_tab = driver.find_element_by_xpath("/html/body/table[6]/tbody/tr/td/form/table/tbody/tr/td/table[1]/tbody/tr/td[3]/a")
        return_tab.click()
        sleep(randint(1,2))


    return
```

```python
def rename_file(old_path,new_name):
    extension = os.path.splitext(old_path)[1]
    dir_name = os.path.dirname(old_path)
    os.rename(old_path,dir_name+'/'+new_name+extension)
    return

def get_last(dir_path):
    return max(glob.iglob(dir_path+'/*'),key=os.path.getctime)
```

```python
api_list = midland_apis[768:]
dir_path = '/Users/alexanderparker/coding/frack-fluid-analysis/data/pdfs'
driver = initialize_driver(dir_path)
index_count = 0
fail_list = []
for api in api_list:
    print(index_count)
    try:
        completion_query(api,driver,dir_path)
    except:
        fail_list.append(api)
    index_count += 1
    
driver.close()
driver.quit()
```
