```python
import sys
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import re
from itertools import chain
import pickle
import glob
```

```python
with open('pickle/midland_big_list.pickle', 'rb') as to_read: #run scrape_data to generate
    midland_big_list = pickle.load(to_read)
```

```python
ua = UserAgent()
user_agent = {'User-agent': ua.random}

def rrc_query(api):
    url_base = 'http://webapps.rrc.texas.gov'
    comp_url = url_base + '/CMPL/publicSearchAction.do'

    request_params = {
        'searchArgs.apiNoHndlr.inputValue' : api,
        'formData.methodHndlr.inputValue' : 'search'

    }

    with requests.post(comp_url, data=request_params, headers=user_agent) as response:
        doc = BeautifulSoup(response.text, 'html.parser')
        return doc
    
def rrc_query_pdf(comp_report):
    url_base = 'http://webapps.rrc.texas.gov'
    comp_url = url_base + '/CMPL/viewPdfReportFormAction.do?method=cmplW2FormPdf&packetSummaryId='

    comp_url = comp_url + comp_report
    
    with requests.post(comp_url, headers=user_agent) as response:
        doc = response.content
        return doc
     

def grab_table(soup):
    table = soup.find("table",{"class":"DataGrid"})
    rows = table.findAll(lambda tag: tag.name=='tr')
    return rows[-1:2:-1]

def grab_comp_id(rows):
    
    id_list = []
    for row in rows:
        id_list.append(re.findall(r'\d+',row.findAll('td')[0].text))
    return list(chain.from_iterable(id_list))
```

```python
comp_report_dict = {}
```

```python
for api in midland_big_list[~midland_big_list['APINumber'].isin(comp_report_dict.keys())]['APINumber']:
    api_query = str(api)[2:10]
    try:
        comp_report_dict[api] = grab_comp_id(grab_table(rrc_query(api_query)))
    except:
        print('Failed to grab ' + str(api))
```

```python
def generate_pdfs(df,dir_name):
    comp_report_dict = {}
    for api in df[~df['APINumber'].isin(comp_report_dict.keys())]['APINumber']:
        api_query = str(api)[2:10]
        try:
            comp_report_dict[api] = grab_comp_id(grab_table(rrc_query(api_query)))
        except:
            print('Failed to grab ' + str(api))
            
    report_list = list(chain.from_iterable(list(comp_report_dict.values())))
            
    dir_path = '/Users/alexanderparker/coding/frack-fluid-analysis/data/pdfs/' + dir_name
    file_list = glob.glob(dir_path+'/*')
    file_list_ids = [re.findall(r'\d+', path)[0] for path in file_list]

    delta_list = list(set(report_list) - set(file_list_ids))

    for report_id in delta_list:
        pdf = rrc_query_pdf(report_id)
        with open(dir_path + '/' + '{}.pdf'.format(report_id), 'wb') as f:
            f.write(pdf)
    return 
```

```python
with open('pickle/midland_all.pickle', 'rb') as to_read: 
     midland_all = pickle.load(to_read)

with open('pickle/delaware_all.pickle', 'rb') as to_read: 
     delaware_all = pickle.load(to_read)
        
with open('pickle/ef_all.pickle', 'rb') as to_read: 
     ef_all = pickle.load(to_read)      
```

```python
generate_pdfs(midland_all,'midland_all')
```
